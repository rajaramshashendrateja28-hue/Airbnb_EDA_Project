TASK 1 importing all the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

TASK 2 getting our data

```python
data=pd.read_csv('datasets.csv',encoding_errors='ignore')
```

TASK 4 Data Cleaning

```python
data.isnull().sum()
```

```python
## dropping all missing values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
# dealing with duplicates
#deleting all the dupliacted rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# type casting
#changing data types

data['id']=data['id'].astype(object)
data['host_id']=data['host_id'].astype(object)
data.dtypes
```

EDA
TASK 5 Data Analysis

Univariate Analysis


```python
#dealing with outliers
df=data[data['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price')
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Availability
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# average price per bed
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
#average price
df.groupby(by='neighbourhood_group')['price'].mean()
```

Feature Engineerring

```python
# Price per bed
df['price_per_bed']=df['price']/df['beds']
df.head()
```

BI variable Analysis
One variable dependency in another variable

```python
df.columns
```

```python
#price dependency on neighbourhood
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
# number of reviews and price relation
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number_of_reviews',y= 'price',hue='neighbourhood_group')
plt.title('Locailty and Review Dependency')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
#Geographical Distribution of Airbnb
plt.figure(figsize=(5,4))
sns.scatterplot(data=df,x='longitude',y= 'latitude',hue='room_type')
plt.title('Geographical Distribution')
plt.show()
```

```python
df.dtypes
```

```python
# Heat map-correlation of one varible with others
corr=df[['latitude','longitude','price','minimum_nights','reviews_per_month','availability_365','beds']].corr()
corr
plt.figure(figsize=(8,8))
sns.heatmap(data=corr,annot=True)
plt.title('correlation')
plt.show()
```
